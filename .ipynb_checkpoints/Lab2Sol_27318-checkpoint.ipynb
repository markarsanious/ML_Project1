## Linear Classification
#### Exercise 2-1:
1. Read file

```python
#Read file and set the 'Student' Column as an index
df = pd.read_csv('TestResults.csv')
df.set_index('Student',inplace=True)
```

```python
#Display number of people who passed VS failed
#print(df['TestResult'].value_counts())

df['TestResult'] = df['TestResult'].map(lambda x: 1 if 'P' in x else -1)
```

2. Define the linear perceptron function

```python
def LinearPerceptron(X,Y,iters,rate):
    
    #Convert DataFrame to Numpy Matrix
    inputMat = np.array(X)
    #Add ones as the input for the bias node
    inputMat = np.c_[np.ones(inputMat.shape[0]),inputMat]
    #Set the Class Column to Numpy Vector
    target = np.array(Y)
    # Initialize Random Weight Vector
    weightVector = np.random.randn(3)

    #Save history of Weights in a Matrix
    weightHistory = weightVector
    #Save history of Errors each Matrix
    errorsHistory = []

    #Loop over the Input Matrix for 'iters' iterations
    for i in range(0,iters):
        # Create error counter
        errors = 0
        #Loop over all the rows in the Input Matrix
        for r in range(0,inputMat.shape[0]):
            #Grab Instance / Input Vector
            instance = inputMat[r]

            #Calculate the Dot Product
            dotProd = sum(instance * weightVector)
            activation = 1 if (dotProd > 0) else -1

            #Increment errors if activation != to expected output
            errors += 1 if (activation!=target[r]) else 0

            #Update Weight Vector when output differs from expected
            weightVector -= rate * (activation-target[r]) * instance

        weightHistory = np.c_[weightHistory, weightVector]
        errorsHistory = np.r_[errorsHistory,errors]

    return weightHistory,errorsHistory
```

3. Train the perceptron

```python
[weightMat,errors] = LinearPerceptron(df[['HoursOfSleep','HoursOfStudy']],df['TestResult'],100,0.4)
```

4. Define helper function for easier plotting

```python
def getLineFromVector(vec,x):
    w0 = vec[0]
    w1 = vec[1]
    w2 = vec[2]
    return ((-1*w0)-(w1*x))/w2
```

5. Plot the results

```python
fig,axes = plt.subplots(2,2)

axes[0][0].scatter(df[df['TestResult']==-1]['HoursOfSleep'], df[df['TestResult']==-1]['HoursOfStudy'], color='red', marker='x')
axes[0][0].scatter(df[df['TestResult']==1]['HoursOfSleep'], df[df['TestResult']==1]['HoursOfStudy'], color='blue', marker='o')
axes[0][0].set_xlabel('Hours Of Sleep')
axes[0][0].set_ylabel('Hours Of Study')
axes[0][0].set_title('Study VS Sleep')
xscale = np.arange(0,round(max(df.max())))
axes[0][0].plot(xscale,getLineFromVector(weightMat[:,(weightMat.shape[1]-1)],xscale),label='line',color='green')

axes[0][1].set_title('Decision Boundaries')
axes[0][1].set_ylim(0,max(xscale))
width = 0.1
for c in range(0,(weightMat.shape[1]-1),2):
    axes[0][1].plot(xscale,getLineFromVector(weightMat[:,c],xscale),lw=width)
    width += 0.1

axes[1][0].plot(np.arange(0,weightMat.shape[1]),weightMat[0,:],label='bias',color='red',ls='-')
axes[1][0].plot(np.arange(0,weightMat.shape[1]),weightMat[1,:],label='w1',color='green',ls='--')
axes[1][0].plot(np.arange(0,weightMat.shape[1]),weightMat[2,:],label='w2',color='blue',ls='-.')
axes[1][0].set_xlabel('Iterations')
axes[1][0].set_ylabel('Weight value')
axes[1][0].set_title('Weights vs Iterations')
axes[1][0].legend(loc=0)

axes[1][1].plot(np.arange(0,errors.shape[0]),errors,color='orange',ls='-')
axes[1][1].set_xlabel('Iterations')
axes[1][1].set_ylabel('Errors')
axes[1][1].set_title('Errors vs Iterations')

plt.tight_layout()

plt.show()
```


## Linear Regression
#### Exercise 2-2:
I. Read the dataset

```python
df = pd.read_csv('USA_Housing.csv')
```

II. Investigate the dataset (see how the data features correlate with each other)

```python
corrmat = df.corr()
corrmat
```

III. Plot the correlation matrix

```python
import seaborn as sns
sns.heatmap(corrmat)
plt.show()
```

IV. Normalize the dataset

```python
# Here we normalize using the vector normalization technique, i.e. if our vector is V we divide V by its magnitude 
# which equals to the square root of the summation of each dimension squared

data_x = np.array(df['Avg. Area Income'])
# print('shape: {0}'.format(data_x.shape))
mag_x = np.sqrt(sum(data_x * data_x))
data_x = data_x/mag_x
data_x = np.c_[np.ones(data_x.shape[0]),data_x]
print(data_x)

data_y = np.array(df['Price'])
mag_y = np.sqrt(sum(data_y * data_y))
data_y = data_y/mag_y
print(data_y)
```

V. Splitting the data into training and testing

```python
order = np.random.permutation(len(data_x))
portion = int(0.8*data_x.shape[0])
test_x = data_x[order[:portion]]
test_y = data_y[order[:portion]]
train_x = data_x[order[portion:]]
train_y = data_y[order[portion:]]
```

VI. Define the gradient calculating function

```python
def get_gradient(w, x, y):
    y_estimate = x.dot(w).flatten()
    error = (y.flatten() - y_estimate)
    mse = (1.0/len(x))*np.sum(np.power(error, 2))
    gradient = -(1.0/len(x)) * error.dot(x)
    return gradient, mse
```

VII. Performing regression

```python
w = np.random.randn(2)
rate = 0.8
tolerance = 1e-6

# Perform Gradient Descent
iterations = 1
while True:
    gradient, error = get_gradient(w, train_x, train_y)
    #print(gradient,error)
    new_w = w - rate * gradient

    # Stopping Condition
    if np.sum(abs(new_w - w)) < tolerance:
        print("Converged")
        break

    # Print error every 50 iterations
    if iterations % 1000 == 0:
        print("Iteration: %dk - MSE: %.8f" % ((iterations/1000), error))

    iterations += 1
    w = new_w

print("Last weight Vector =", w)
print("Test Cost =", get_gradient(w, test_x, test_y)[1])
```

VIII. Plot the results

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])

ax.plot(data_x[:,1], data_x.dot(w), c='g', label='Model')
ax.scatter(train_x[:,1], train_y, c='b', label='Train Set')
ax.scatter(test_x[:,1], test_y, c='r', label='Test Set')
ax.grid()
ax.legend(loc='best')
ax.set_xlabel('Average Income')
ax.set_ylabel('Price')
ax.set_title('Price VS Avg. Income')

plt.show()
```

### Linear regression using more features and sklearn
#### Exercise 2-3
1. Imports needed

```python
from sklearn import preprocessing
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

2. Read data

```python
#Read Data
df = pd.read_csv('USA_Housing.csv')
# X = df[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms','Avg. Area Number of Bedrooms','Area Population']]
X = df['Avg. Area Income'].reshape(-1,1)
Y = df['Price']
```

3. Normalize Data

```python
#Normalize Data
X_scaled = preprocessing.scale(X)
Y_scaled = preprocessing.scale(Y)
```

4. Split data into training and testing sets

```python
#Split data into Train/Test
X_train,X_test,Y_train,Y_test = train_test_split(X_scaled,Y_scaled,test_size=0.2,random_state=1)
```

5. Fit the model

```python
#Model fitting
lm = LinearRegression()
lm.fit(X_train,Y_train)
```

6. Display the coeffecients and intercept

```python
#Coeffecients
coeffecients = lm.coef_
intercept = lm.intercept_
```

7. Use the model to predict the price and calculate the mean and squared absolute errors

```python
#Check Mean Absolute Error
predictions = lm.predict(X_test)
mae = metrics.mean_absolute_error(predictions,Y_test)
print(mae)
mse = metrics.mean_squared_error(predictions,Y_test)
print(mse)
```
